```python
import pandas as pd
```

```python
df1 = pd.read_csv('datasets/init/malicious_68k.csv')
df2 = pd.read_csv('datasets/init/benign_68k.csv')

if 'Unnamed: 0' in df1.columns:
    df1.drop(['Unnamed: 0'], axis=1, inplace=True)

if 'Unnamed: 0' in df2.columns:
    df2.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df1
```

```python
df2
```

```python
df_concat = pd.concat([df1, df2])

df_concat
```

```python
df_concat = df_concat.sample(frac=1, random_state=42).reset_index(drop=True)

df_concat.to_csv('datasets/intermediate/shuffled_136k.csv', index=False)
```
